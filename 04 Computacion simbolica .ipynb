## polinomios de Chebyshev

De primer tipo

1. $T_{0}(x)=1$
2. $T_{1}(x)=x$
3. $T_{n+1}(x)=2xT_{n}(x)-T_{n-1}(x)$

De segundo tipo

1. $U_{0}(x)=1$
2. $U_{1}(x)=2x$
3. $U_{n+1}(x)=2xU_{n}(x)-U_{n-1}(x)$

```python
x=sp.Symbol('x')
```

```python
t0=1
t1=x
```

```python
t2=2*x*t1-t0
t2
```

```python
t3=2*x*t2-t1
t3.expand()
```

```python
t4=2*x*t3-t2
t4.expand()
```

```python
t4
```

```python
def chebyshev_tipouno(n=2):
    polinomios=[1,x]
    for i in range(2,n+1):
        polinomios.append(2*x*polinomios[i-1]-polinomios[i-2])
    return polinomios
```

```python
poly=chebyshev_tipouno(50)
len(poly)
```

```python
poly[20]
```

```python
poly[20].expand
```

```python
#expandir 
def chebyshev_Conex(n=2):
    polinomios=[1,x]
    for i in range(2,n+1):
        thispoly=2*x*polinomios[i-1]-polinomios[i-2]
        polinomios.append(thispoly.expand())
    return polinomios
```

```python
poly=chebyshev_Conex(50)
len(poly)
```
